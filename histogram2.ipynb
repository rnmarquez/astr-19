```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng = np.random.default_rng()
x = rng.random(1001) 

plt.hist(x, bins=100, color='skyblue', edgecolor='black')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Histogram Plot')
plt.savefig('histogram_plot.pdf')

plt.show()
```
