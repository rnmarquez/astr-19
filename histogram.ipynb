```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
random_numbers = np.random.uniform(0, 1, 1000)

plt.hist(random_numbers, bins=100, color='skyblue', edgecolor='black')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Histogram Plot')
plt.savefig('histogram_plot.pdf')
plt.show()
```
