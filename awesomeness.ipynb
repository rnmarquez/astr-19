```python
import matplotlib.pyplot as plt
import math 
```

```python
#Creating x value array between 0-1 with steps of 100
x_arr = []
for i in range(101):
    x_arr.append(i/100)

#Creating the y=exp(x) function with each value inside our x arr
y_arr = []
for num in x_arr:
    y_arr.append(math.exp(num))
    
#Plotting the graph about my awesomness
plt.plot(x_arr,y_arr)
plt.xlabel("Time (milliseconds)")
plt.ylabel("Awesomeness")
plt.title('My Awesome Graph')
plt.show
plt.savefig('Awesomeness.pdf', bbox_inches='tight',dpi=600)
```
