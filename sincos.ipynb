```python
import matplotlib.pyplot as plt
import math
```

```python
x_arr = []
for i in range(101):
    x_arr.append(i/100)

sin_arr = []
cos_arr = []

for num in x_arr:
    sin_arr.append(math.sin(num))
    cos_arr.append(math.cos(num))

f, axarr = plt.subplots(1,2)

axarr[0].plot(x_arr,sin_arr)
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')
axarr[0].set_title('Sin Graph')

axarr[1].plot(x_arr,cos_arr)
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')
axarr[1].set_title('Cos Graph')

plt.show()
```
